```python
#Bibliotecas
import urllib.request as urllib_request
import bs4
import pandas as pd

print('BeautfulSoup: ' + bs4.__version__)
print('urllib3: ' + urllib_request.__version__)
print('Pandas:' + pd.__version__)
```

```python
# ACESSANDO A URL
from urllib.request import urlopen, Request
from urllib.error import URLError, HTTPError
from bs4 import BeautifulSoup

url = 'https://towardsdatascience.com/4-libraries-that-can-perform-eda-in-one-line-of-python-code-b13938a06ae'

headers = {'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'}
user = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'

def trata_html(input):
    return " ".join(input.split()).replace('> <','><')

def extrae_texto(input):
    sem_repeticao = []
    i = 0
    
    for tag in input.findAll('body'):
        if i <= 1:
            sem_repeticao.append(tag.getText())
        else:
            if tag.getText() != '' and tag.getText() != sem_repeticao[-1]:
                sem_repeticao.append(tag.getText())
        i += 1

    return sem_repeticao

try:
    req = Request(url,headers=headers)  # fazendo um requisição ao servidor como um usuário
    response = urlopen(req)             # abringo a url
    html = response.read()              # lendo o conteúdo do response
    print('O objeto html é do tipo: ' + str(type(html)))                   # o conteúdo html esta como byte
    html = html.decode('utf-8-sig')         # convertendo o conteúdo html para string
    print('O objeto html convertido para: ' + str(type(html)))
    #print(html)
    print('O código html é tratado.')
    html = trata_html(html)             # Limpa o conteúdo html para melhor tratamento com o BeautifulSoup
    #print(html)
    html = BeautifulSoup(html,'html.parser')    # Cria um objeto BeautifulSoup
    print('O objeto html é convertido em: ' + str(type(html)))
    #print(html.prettify())
    print('Extraido apenas o texto do arquivo html.')
    texto = extrae_texto(html)
    #print(texto)
    print('Após o tratamento ficamos com: ' + str(len(texto)) + ' linhas.')
except HTTPError as e:
    print(e.status, e.reason)
except URLError as e:
    print(e.reason)
```

# Tokenização

```python
# Bibliotecas
import nltk
#nltk.download('punkt')
print('nltk: ' + nltk.__version__)
```

```python
def separa_palavras(lista_tokens):
    lista_palavras = []
    for token in lista_tokens:
        if token.isalpha():
            lista_palavras.append(token)
    return lista_palavras

def normalizacao(lista_palavras):
    lista_normalizada = []
    for palavra in lista_palavras:
        lista_normalizada.append(palavra.lower())
    return lista_normalizada

palavras_separadas = nltk.tokenize.word_tokenize(str(texto))
print('Após a tokenização temos: ' + str(len(palavras_separadas)) + ' palavras')
palavras_separadas = separa_palavras(palavras_separadas)
print('Após a limpeza de caracteres especiais temos: ' + str(len(palavras_separadas)) + ' tokens.')
palavras_separadas = normalizacao(palavras_separadas)
palavras_sem_repeticao = set(palavras_separadas)
print('Após a normalização temos: ' + str(len(palavras_sem_repeticao)) + ' tokens.')
```
